**Poke API**

```python
import requests
```

```python
base_url = "https://pokeapi.co/api/v2/"

#lấy url theo tên từng pokemon: base_url/pokemon/tên pokemon
def get_pokemon_url(name):
    pokemon_url = f"{base_url}/pokemon/{name}"
    response = requests.get(pokemon_url)

    if response.status_code ==200:
        print("Lay data ok")
        data = response.json()
        return data
    else:
        print(f"hong lay duoc data, statuscode ={response.status_code}")

pikachu_data = get_pokemon_url("pikachu")
```
